```python
import pandas as pd
```

```python
df=pd.read_excel("dataset.xlsx",index_col=0)
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder()
ohe.fit(df[['marque']])
```

```python
ohe.categories_[0]
```

```python
marques_sparse=ohe.transform(df[['marque']])
marques=marques_sparse.toarray()
marques
```

```python
df_marque=pd.DataFrame(marques,index=df.index,columns='marque_'+ohe.categories_[0])
df_marque
```

```python
df=pd.concat([df,df_marque], axis=1)
df
```

```python
df.drop('marque', axis=1, errors='ignore', inplace=True)
df
```
